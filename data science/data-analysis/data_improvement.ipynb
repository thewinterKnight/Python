```python
df_clean = df.apply(lambda x : x.fillna(x.value_counts().index[0]))
df_clean
```

```python
df_clean.isnull().sum()
```

```python
df_clean['Date of Birth'] = pd.to_datetime(df_clean['Date of Birth'], format='%d/%m/%Y')
df_clean
```

```python
df_clean.info()
```

```python
df_clean['year'] = df_clean['Date of Birth'].dt.year
df_clean['month'] = df_clean['Date of Birth'].dt.month
df_clean['day'] = df_clean['Date of Birth'].dt.day
df_clean
```

```python
df_clean.info()
```

```python
season_data = df.groupby('Season')
for key, value in season_data:
    print('*{}*, {}\n\n'.format(key, season_data.get_group(key)))

df_clean.groupby('Season').first()
```

```python
plt.figure(figsize=(10, 6))
sns.jointplot(x='Season', y='Name', data=df_clean)
```

```python
df_clean.rename(columns = {'Name': 'name', 'Date of Birth': 'dob', 'Food Preference': 'foodpreference', 'Height': 'height', 'Weight': 'weight', 'Last_Modified': 'lastmodified', 'Season': 'season'}, inplace=True)
df_clean
```

```python
df_clean['season'].value_counts()
```

```python
data_dummy = pd.get_dummies(df_clean[['foodpreference', 'season']], drop_first=False)
data_dummy
```

```python
df_clean = pd.concat([df_clean, data_dummy], axis=1)
df_clean
```

```python
df_clean = df_clean.drop(['dob', 'foodpreference', 'season'], axis=1)
df_clean
```

```python
df_clean.lastmodified.unique()
```

```python
df_clean.loc[:, 'lastmodified'] = df_clean['lastmodified'].apply(lambda x: '2016' if x=='<2017' else x)
df_clean
```

```python
df_clean['day_sin'] = np.sin(df_clean.day*(2.*np.pi/31))
df_clean['day_cos'] = np.cos(df_clean.day*(2.*np.pi/31))
df_clean['month_sin'] = np.sin(df_clean.month*(2.*np.pi/12))
df_clean['month_cos'] = np.cos(df_clean.month*(2.*np.pi/12))

df = df_clean.drop(['day', 'month', 'year'], axis=1)
df
```
